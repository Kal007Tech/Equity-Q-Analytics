```python
# Import libraries
import pandas as pd
import numpy as np
import yfinance as yf
import matplotlib.pyplot as plt
```

# Downloading and Extracting Tickers for Information Technology and Utilities Sector in S&P 500

```python
# Download the S&P 500 company data from Wikipedia
sp500_df = pd.read_html('https://en.wikipedia.org/wiki/List_of_S%26P_500_companies')[0]

# Filter the DataFrame to only include companies in the Information Technology sector
IT_sp500_df = sp500_df[sp500_df['GICS Sector'] == 'Information Technology']

# Sort the filtered DataFrame by ticker symbol in alphabetical order
sorted_IT_sp500_df = IT_sp500_df.sort_values('Symbol')

# Extract the list of ticker symbols from the sorted DataFrame
IT_sp500_tickers = sorted_IT_sp500_df['Symbol'].tolist()

# Print the list of ticker symbols and number of stocks in Information Technology sector
print(IT_sp500_tickers)
len(IT_sp500_tickers)
```

```python
# Download the S&P 500 company data from Wikipedia
sp500_df = pd.read_html('https://en.wikipedia.org/wiki/List_of_S%26P_500_companies')[0]

# Filter the DataFrame to only include companies in the Utilities sector
utilities_sp500_df = sp500_df[sp500_df['GICS Sector'] == 'Utilities']

# Sort the filtered DataFrame by ticker symbol in alphabetical order
sorted_utilities_sp500_df = utilities_sp500_df.sort_values('Symbol')

# Extract the list of ticker symbols from the sorted DataFrame
utilities_sp500_tickers = sorted_utilities_sp500_df['Symbol'].tolist()

# Print the list of ticker symbols and number of stocks in Utilities sector
print(utilities_sp500_tickers)
len(utilities_sp500_tickers)
```

# Downloading Stock Price Data and Calculating Daily Log-Returns,Visualization

# Information Technology Sector

```python
#Import stock prices from Yahoo Finance:
IT_adj_close = yf.download(IT_sp500_tickers, start = "2022-01-01", end = "2022-12-31")["Adj Close"]
IT_adj_close
```

```python
#Calculate the daily log-returns for each stock
IT_log_returns = np.log(IT_adj_close / IT_adj_close.shift(1))
print(IT_log_returns)
```

```python
#Calculate annualized log-return
IT_annualized_returns = ((1 + IT_log_returns.mean()) ** 252 - 1)*100

#Create a DataFrame
IT_df = pd.DataFrame({'Annualized Returns': IT_annualized_returns})

# Round values to two decimal places
IT_df['Annualized Returns'] = IT_df['Annualized Returns'].round(2)

#Sort Data
IT_df_sorted = IT_df.sort_values(by='Annualized Returns', ascending=False)

print(IT_df_sorted)
```

```python
# Visualize the annualized returns using a bar chart
IT_df_sorted.plot(kind='bar', figsize=(20,10))
plt.title('Performance of Information Technology Sector', fontsize=18)
plt.ylabel('Annualized Returns (%)', fontsize=14)
plt.xlabel('Tickers', fontsize=14)
plt.show()
```

# Annualized Returns for Energy stocks in descending order

```python
import numpy as np
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt

# Define the list of tickers
tickers = ['APA', 'BKR', 'COP', 'CTRA', 'CVX', 'DVN', 'EOG', 'EQT', 'FANG', 'HAL', 'HES', 'KMI', 'MPC', 'MRO', 'OKE', 'OXY', 'PSX', 'PXD', 'SLB', 'TRGP', 'VLO', 'WMB', 'XOM']
# Define the start and end dates
start_date = '2022-01-01'
end_date = '2022-12-31'

# Getting the price data from yahoo 
energy_adj_close = yf.download(tickers, start=start_date, end=end_date)["Adj Close"]

# Calculating the daily log returns for each stock
log_returns = np.log(energy_adj_close / energy_adj_close.shift(1))
print(log_returns*100)
```

```python
#Calculate annual log-return
annualized_returns = ((1 + log_returns.mean()) ** 252 - 1)*100

#Creating a df
energy_df = pd.DataFrame({'Annualized Returns': annualized_returns})
energy_df["Annualized Returns"] = energy_df["Annualized Returns"].round(2)
energy_df_sorted = energy_df.sort_values(by = "Annualized Returns", ascending = False)
print(energy_df_sorted)

# Visualize the holding period returns using a bar chart
energy_df_sorted.plot(kind='bar', figsize=(10,10))
plt.title('Annualized Returns for Energy Stocks')
plt.ylabel('Return (%)')
plt.show()
```

# Correlation Performance of S&P 500 sectors in 2022

```python
import pandas as pd
import numpy as np
import yfinance as yf
import seaborn as sns
import matplotlib.pyplot as plt

# Define S&P500 sectors
sectors = {
    'S&P500': ['^GSPC'],
    'Communication Services': ['XLC'],
    'Consumer Discretionary': ['XLY'],
    'Consumer Staples': ['XLP'],
    'Energy': ['XLE'],
    'Financials': ['XLF'],
    'Health Care': ['XLV'],
    'Industrials': ['XLI'],
    'Information Technology': ['XLK'],
    'Materials': ['XLB'],
    'Real Estate': ['XLRE'],
    'Utilities': ['XLU']
}

# Download historical data for S&P500 stocks
data = yf.download(list(np.concatenate(list(sectors.values()))), start='2022-01-01')['Adj Close']

# Calculate daily returns
returns = data.pct_change()

# Calculate correlation matrix
corr_matrix = returns.corr()

# Plot correlation matrix
sns.set(font_scale=1.2)
plt.figure(figsize=(8, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title('S&P500 Correlation Matrix')
plt.show()
plt.savefig("S&P500 Correlation Matrix.pdf")
```

# Comparison of Annualized Return of S&P 500 and its Sectors

```python
import pandas as pd
import yfinance as yf
import matplotlib.pyplot as plt

# Define tickers and sector ETFs
tickers = ['^GSPC', 'XLC', 'XLY', 'XLP', 'XLE', 'XLF', 'XLV', 'XLI', 'XLB', 'XLRE', 'XLK', 'XLU']
sectors = ['S&P500', 'Consumer Discretionary', 'Consumer Staples', 'Energy',
           'Financials', 'Health Care', 'Industrials', 'Materials', 'Real Estate', 'Technology', 'Utilities']

# Download historical data for each ticker
data = pd.DataFrame()
for ticker in tickers:
    try:
        ticker_data = yf.download(ticker, start='2022-01-01', end='2022-12-31')['Adj Close']
        ticker_data.name = ticker
        data = pd.concat([data, ticker_data], axis=1)
    except KeyError:
        print(f"'{ticker}' data not found")

# Calculate daily returns and annualize them
returns = data.pct_change()
annualized_returns = ((1 + returns.mean()) ** 252 - 1) * 100

# Create a DataFrame of annualized returns for each sector
sector_returns = pd.DataFrame({'Sector': sectors, 'Return': [annualized_returns[tickers[i]] for i in range(len(sectors))]})
sector_returns = sector_returns.sort_values('Return', ascending=False).reset_index(drop=True)

# Set the color for positive and negative annualized returns
positive_color = 'green'
negative_color = 'red'

# Plot the holding period return of each sector and the index in a bar chart
plt.figure(figsize=(12,6))
plt.bar(sector_returns['Sector'], sector_returns['Return'], color=[positive_color if r >= 0 else negative_color for r in sector_returns['Return']])
plt.axhline(y=0, color='black', linestyle='-')
plt.title('Annualized Return of S&P 500 and Its Sectors', fontsize=16)
plt.xlabel('Sector', fontsize=14)
plt.ylabel('Annualized Return (%)', fontsize=14)
plt.xticks(rotation=90)
plt.show()
```
